3. In Tableau Server or Cloud (conceptually), which feature allows administrators to control which users
can view, interact with, or edit a published dashboard?

##ans.
In **Tableau Server** or **Tableau Cloud**, the feature that allows administrators to control which users can **view, interact with, or edit** a published dashboard is called **Permissions** (also known as **User Permissions and Roles**).

### How Permissions Work

Administrators assign permissions to **users** or **groups** to define what actions they are allowed to perform on dashboards, workbooks, projects, and data sources.

Common permission levels include:

* **View** – Users can open and view the dashboard.
* **Interact** – Users can filter, sort, drill down, and use interactive dashboard features.
* **Download** – Users can download the workbook, data, or PDF (if permitted).
* **Edit (Web Edit)** – Users can modify the workbook directly in the browser.
* **Publish/Save** – Users can publish or save changes to the server.
* **Delete** – Users can remove content if they have the required permissions.

### Role of User Roles

In addition to permissions, **site roles** determine the maximum capabilities a user can have. For example:

* **Viewer** – Can only view and interact with dashboards.
* **Explorer** – Can interact, create, and edit content (depending on permissions).
* **Creator** – Has full authoring and publishing capabilities.

### Answer (Exam Point of View)

**Feature:** **Permissions (User Permissions and Roles)**

**Purpose:** It enables administrators to control which users or groups can **view, interact with, edit, publish, or manage** published dashboards and other Tableau content, ensuring proper security and access control.


4. If you want users to view a dashboard but not download the underlying data, which permission setting
would you adjust in Tableau Server

##ans.

To allow users to **view a dashboard but prevent them from downloading the underlying data**, you should adjust the **Permissions** for the workbook or view in **Tableau Server/Cloud**.

Specifically:

* **Allow:** **View**
* **Deny (or leave ungranted):** **Download Data** (sometimes shown as **Download Full Data** or **Download Summary Data**, depending on the version and settings)

This lets users open and interact with the dashboard while preventing them from downloading the data behind the visualizations.

### Exam Answer

**Permission to adjust:** **Download Data** permission.

**Configuration:**

* ✅ **View** = Allowed
* ❌ **Download Data** = Denied/Not Allowed

This ensures users can **view and interact with the dashboard** but **cannot download the underlying data**.



5. Name two methods by which a Tableau dashboard can be embedded into a web page or website.

##ans.
Two common methods to embed a **Tableau dashboard** into a web page or website are:

1. **Embed Code (HTML `<iframe>`)**

   * Tableau provides an **Embed Code** when a dashboard is published.
   * You can copy and paste this HTML code (typically an `<iframe>`) into your website's HTML to display the dashboard.

2. **Tableau Embedding API (JavaScript API)**

   * Use the **Tableau Embedding API** to embed dashboards with JavaScript.
   * This method allows greater customization, such as controlling filters, parameters, events, and interactions programmatically.

### Exam Answer

* **Method 1:** Embed the dashboard using the **HTML Embed Code (`<iframe>`)**.
* **Method 2:** Embed the dashboard using the **Tableau Embedding API (JavaScript API)** for advanced interactive integration.


6. List two best practices you should follow when preparing a dashboard for public sharing to ensure data
security and privacy.

##ans.
Two important best practices for preparing a Tableau dashboard for public sharing are:

1. **Remove or Protect Sensitive Data**

   * Ensure the dashboard does **not contain confidential or personally identifiable information (PII)** such as names, phone numbers, email addresses, financial records, or passwords.
   * Use aggregated or anonymized data whenever possible.

2. **Configure Permissions and Access Properly**

   * Set appropriate **permissions** so users can only access the information they are authorized to see.
   * Disable options such as **Download Data** if users should not be able to access the underlying data.

### Exam Answer

* **Remove or anonymize sensitive/confidential data** before publishing.
* **Configure appropriate permissions and restrict data downloads** to prevent unauthorized access.


7.After publishing a dashboard, users report a “Data not found” error.
 What is the most common cause related to data source settings?


 #ans.
 The **most common cause** is that the **published data source connection is broken or incorrectly configured**.

This can happen if:

* The database or file path has changed.
* The data source was not published along with the workbook.
* The server cannot access the database or network location.
* Database credentials are missing, expired, or incorrect.

### Exam Answer

**Cause:** The dashboard's **data source connection is broken or misconfigured** (e.g., invalid file path, inaccessible database, or incorrect credentials), causing Tableau Server/Cloud to be unable to locate the data.

**Solution:** Verify the data source connection, update the connection details or file path, ensure the data source is published and accessible, and confirm that the correct database credentials are configured.



8. When embedding a Tableau dashboard into a secure internal portal, what two additional security
measures should be implemented to ensure only authorized users can access the embedded content

##ans.
When embedding a Tableau dashboard into a **secure internal portal**, implement these two security measures:

1. **User Authentication (Single Sign-On or Login)**

   * Require users to authenticate before accessing the embedded dashboard.
   * Common methods include **Single Sign-On (SSO)** using SAML, OpenID Connect, or Active Directory, ensuring users are verified.

2. **Role-Based Permissions (Authorization)**

   * Configure **Tableau permissions** so only authorized users or groups can view or interact with the dashboard.
   * Grant access based on user roles and follow the principle of least privilege.

### Exam Answer

1. **Enable user authentication (e.g., Single Sign-On or secure login)** to verify user identity.
2. **Apply Tableau user/group permissions (role-based access control)** so only authorized users can access the embedded dashboard and its data.
